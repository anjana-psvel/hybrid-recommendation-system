```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
movies= pd.read_csv('tmdb_5000_movies.csv')
credits= pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
print(movies.shape)
print(credits.shape)
```

```python
# prompt: rename the movie_id into id
credits = credits.rename(columns={'movie_id': 'id'})
credits.head()
```

```python
movies_credits = pd.merge(movies, credits, on='id')
movies_credits = movies_credits.drop(columns=['homepage', 'status', "production_countries","title_x", "title_y"])
movies_credits.head()
```

calculate weightted average  :
W = (R*v + C*m ) / v + m

```python
movies_credits.columns
```

```python
R = movies_credits['vote_average']
v = movies_credits['vote_count']
C = movies_credits['vote_average'].mean()
m = movies_credits['vote_count'].quantile(0.7)
```

```python
movies_credits["weighted average"]= ((R*v) + (C*m))/(v+m)
```

```python
movies_credits.head()
```

```python
# sort based on weighted avg

movies_credits = movies_credits.sort_values('weighted average', ascending=False)
movies_credits.head()
```

```python
movies_credits.columns
```

```python
movies_credits[["id", "original_title","vote_count", "vote_average", "weighted average", "popularity"]].head(20)
```

```python
# rank the dataset based on popularity

movies_credits_ranked = movies_credits.sort_values('popularity', ascending=False)
movies_credits_ranked[["id", "original_title", "weighted average", "popularity"]].head(20)
```

Recommend based on 50% popularity and 50% weighted average

```python
# import minmax scaler

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
movies_credits_scaled = scaler.fit_transform(movies_credits[["weighted average", "popularity"]])
movies_normalized = pd.DataFrame(movies_credits_scaled, columns=["weighted average", "popularity"])
movies_normalized.head()
```

```python
movies_credits[["normalized_WA", "normalized_popularity"]] = movies_normalized
movies_credits.head()
```

```python
movies_credits["score"]= movies_credits["normalized_WA"] * 0.5 + movies_credits["normalized_popularity"] * 0.5
movies_normalized= movies_credits.sort_values(["score"], ascending=False)
movies_normalized[['id', 'original_title', 'score', 'normalized_WA', 'normalized_popularity']].head(20)
```
